```python
import tweepy
from tweepy import OAuthHandler

consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''

auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
# searches = ['16Personalities', 'AnthonyKusuma']
searches = ['ENFJ', 'ENFP', 'ENTJ', 'ENTP', 'ESFJ', 'ESFP', 'ESTJ', 'ESTP', 'INFJ', 'INFP', 'INTJ', 'INTP', 'ISFJ', 'ISFP', 'ISTJ', 'ISTP']

for search in searches:
  print(search)
  for tweet in api.search(q=search, tweet_mode="extended", lang='in'):
    if hasattr(tweet, "retweeted_status"):
      continue
    else:
      print(tweet.id_str, tweet.user.screen_name)
      print(tweet.full_text)
      print('------------------------------------------------------------------------')
```
